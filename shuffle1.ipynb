```python
import ray
ray.init(ignore_reinit_error=True,num_cpus=10)
```

```python
categories = []
file = "categories.txt"
with open(file, 'r') as f:
  for index,category in enumerate(f):
    category = category.strip()
    category_pdfs = category
    categories.append(category_pdfs)
with open(file, 'w') as f:
  for category in categories:
    f.write(category+"**.pdf\n")
print(categories)
```

```python
!rm -rf /content/2*
!rm -rf /content/sample_data/
```

```python
import subprocess
import random
```

```python
partitions = [(category.strip()).split("/")[-2] for category in categories]
category_list=[]
category_list = categories
```

```python
@ray.remote
def upload_papers(path,partition,category):
  try:
    print(path+partition)
    subprocess.run(["mkdir",path+"/"+partition])
    subprocess.run(["gsutil", "cp", "-n", category+"**.pdf", path+partition],check=True)
  except Exception as e:
    print(e)
```

```python
zipped_list = zip(partitions,category_list)
zipped = [(x, y) for x, y in zipped_list]
random.shuffle(zipped)
print(zipped)
remotes = []
for item in zipped:
  partition , category = item
  remote_id = upload_papers.remote(path="/content/drive/MyDrive/paper_archive/",partition=partition,category=category)
  remotes.append(remote_id)
ray.get(remotes)
```
